# Implement function to find limit :

```python
def limx(x,func,del_x = 0.000001,rounding_fig=3):
  '''
    There is no set of values for del_x and rounding_fig for which we can determine
    all possible values of left_side_limit and right_side_limit.
    If we make del_x as a function of input then we ran into problem of underflow.
    If value will not in domain then it will give error.
  '''
  try:
    func(x)
  except:
    print("Value is not in function domain")
    return
  else:
    lhl = round(func(x+del_x),rounding_fig)
    rhl = round(func(x-del_x),rounding_fig)
    if(lhl==rhl):
      print('Limit Exists at ',x)
    else:
      print('Limit Does not Exist at ',x)

def one_by_x(x):
  return 1/x

limx(100,one_by_x)
```

```python
limx(0,one_by_x)
```

```python
import math as m
limx(1,m.log)
```

# Implementing Function to find derivative :

```python
def derx(x,func,del_x = 0.000001,rounding_fig = 6):
  '''
    There is no set of values for del_x and rounding_fig for which we can determine
    all possible values of left_side_limit and right_side_limit.
    If we make del_x as a function of input then we ran into problem of underflow.
    If value will not in domain then it will give error.
  '''
  try:
    func(x)
  except:
    print("Value is not in function domain")
    return
  else:
    lhs = round(func(x+del_x)-func(x),rounding_fig)/del_x
    rhs = round(func(x-del_x)-func(x),rounding_fig)/(-del_x)
    if(lhs==rhs):
      print('derivative of x at ',x,' is ',lhs)
    else:
      print('derivative does not exist at x = ',x)

def x_sqr(x):
  return x**2

def modx(x):
  return abs(x)

derx(0,modx)
```

```python
def inv_modx(a):
  return 1/abs(a)

derx(0,inv_modx)
```

# Implementing Gradient Descent From Scratch

```python
def function(x):
  return x**2-4

def derv_func(x):
  return float(2*x)
```

```python
import numpy as np

def gradient_descent(n=0.1,patience=5):
  x_old = np.random.randint(0,100)
  dy_dx0 = derv_func(x_old)
  i = 1
  pat = 0
  while ((abs(dy_dx0) > 0.001) and (pat<patience)):
    x_new = round(x_old - n*dy_dx0,2)
    print('Iteration No : ',i,' x_old : ',x_old,' x_new : ',x_new,' dy_dx : ',dy_dx0)
    print('-'*50)
    dy_dx = dy_dx0
    dy_dx0 = derv_func(x_new)
    x_old = x_new
    if(dy_dx==dy_dx0):
      pat+=1
    i+=1
  print('Function is minimum at x : ',x_old)
```

```python
gradient_descent()
```

```python
#Importing Libraries
import pandas as pd
import numpy as np
import scipy.linalg as la
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
#Reading CSV
df = pd.read_csv("./mnist.csv")
df.head()
```

```python
#Dropping label and storing in x
x = df.drop(['label'],axis=1)
```

```python
x.head()
```

```python
#Taking input of a value an printing it's corresponding image and label
a = int(input("Enter a value : "))
plt.imshow(np.reshape(x.loc[a].to_numpy(),(28,28)),cmap='gray')
print('Label : ',df['label'][a])
```

```python
ss = StandardScaler()
x_std = ss.fit_transform(x)
```

```python
cov_mat = (np.transpose(x_std))@(x_std)
```

```python
values, vectors = la.eigh(cov_mat, eigvals=(782,783))
```

```python
values
```

```python
vectors.shape
```

```python
# converting the eigen vectors into (2,d) shape for ease of computation which we do
vector = vectors.T
vector[[0,1]]=vector[[1,0]]
```

```python
# projecting the original data onto the eigen basis.
import matplotlib.pyplot as plt
new_coordinates = np.matmul(vector, x.T)
```

```python
print (" resultant new data points' shape ", vector.shape, "X", x.T.shape)
```

```python
new_coordinates = np.vstack((new_coordinates, df['label'].to_numpy())).T
```

```python
new_coordinates
```

```python
dataframe = pd.DataFrame(data=new_coordinates, columns=("1st_principal", "2nd_principal","label"))
print(dataframe.head())
```

```python
import seaborn as sn
sn.FacetGrid(dataframe, hue="label",height=8).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```
